```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv('heart.csv')
```

```python
# Exploring the data
print(df.head()) # view the first 5 rows
print(df.shape) # view the dimensions of the dataframe
print(df.columns) # view the column names
print(df.info()) # view information about the data
```

```python
# Cleaning the data
df = df.drop_duplicates() # remove duplicated rows
df = df.dropna() # remove rows with missing values
```

```python
# Visualizing the data
sns.countplot(x='HeartDisease', data=df)
plt.title('Distribution of Heart Disease')
plt.show()

sns.boxplot(x='HeartDisease', y='Age', data=df)
plt.title('Age Distribution by Heart Disease')
plt.show()

sns.boxplot(x='HeartDisease', y='Cholesterol', data=df)
plt.title('Cholesterol Distribution by Heart Disease')
plt.show()

sns.scatterplot(x='MaxHR', y='Oldpeak', hue='HeartDisease', data=df)
plt.title('MaxHR vs Oldpeak by Heart Disease')
plt.show()

```

```python
# Creating dummy variables for categorical variables
df = pd.get_dummies(df, columns=['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope'])

#Dropping the output variable for correlation matrix
df = df.drop(['HeartDisease'], axis=1)

# Creating the correlation matrix
corr_matrix = df.corr()

# Visualizing the correlation matrix
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap=sns.diverging_palette(20, 220, n=200), vmin=-1, vmax=1, center=0, square=True, linewidths=0.5, cbar_kws={"shrink": .5}, fmt='.2f', annot_kws={"size": 10})
plt.title('Correlation Matrix')
plt.show()
```
